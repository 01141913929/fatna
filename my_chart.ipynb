```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. إنشاء بيانات بسيطة باستخدام Pandas و NumPy
data = {
    'المدن': ['القاهرة', 'الأقصر', 'الإسكندرية', 'أسوان'],
    'درجة الحرارة': np.array([35, 42, 32, 44]) # استخدمنا NumPy لإنشاء مصفوفة الأرقام
}
df = pd.DataFrame(data)

print("--- جدول البيانات الذي أنشأناه ---")
print(df)
print("---------------------------------")


# 2. رسم البيانات باستخدام Matplotlib
plt.figure(figsize=(8, 6)) # لتحديد حجم الرسمة

plt.bar(df['المدن'], df['درجة الحرارة'], color=['#FF5733', '#C70039', '#900C3F', '#581845'])
plt.title('درجات الحرارة الحالية في بعض المدن المصرية')
plt.xlabel('المدينة')
plt.ylabel('درجة الحرارة (مئوية)')
plt.show() # لإظهار الرسم البياني
```
